```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

Popular NFTs

```python

df['asset.name'].unique()#Check for the assets involved
```

```python
len(df['asset.name'].unique())#total number of unique assets
```

```python
df.groupby('asset.name').size()#return count of each unique asset name
```

```python
assetname_counts=df.groupby('asset.name').size()#creating new variable
assetname_counts
```

```python
assetname_counts=assetname_counts.to_frame()#convert data into dataframe
assetname_counts
```

```python
assetname_counts=assetname_counts.rename(columns={0:'count'})#Change name of count from 0 to count
assetname_counts
```

```python
assetname_counts=assetname_counts.sort_values(by='count',ascending=False)#sort values in descending form
assetname_counts.head(20)
```

```python
assetname_counts.columns
```

```python
assetname_counts=assetname_counts[assetname_counts.index != 'unknown']
assetname_counts
```

```python
#Most Common Assets
fig,ax=plt.subplots()
ax.bar(assetname_counts.index[:20],assetname_counts['count'][:20])
ax.set_title('Most Popular sold NFT Assets')
ax.set_ylabel('Number of Assets Sold')
ax.set_xticklabels(labels=assetname_counts[:20].index,rotation=90)
plt.show()
```
